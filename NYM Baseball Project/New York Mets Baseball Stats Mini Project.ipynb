```python
# Creates a new dataframe with the average age from 2022 and 2023 batters
df_batting_Age = [df_batting_2023['Age'].mean(), df_batting_2022['Age'].mean()]

# Creates a new dataframe with the total hits from 2022 and 2023 batters
df_batting_hits = [df_batting_2023['Hits'].sum(), df_batting_2022['Hits'].sum()]

# Creates a new dataframe with the total home runs from 2022 and 2023 batters
df_batting_HomeRuns = [df_batting_2023['Home_Runs'].sum(), df_batting_2022['Home_Runs'].sum()]

# Creates a new dataframe with the total times hit by pitch in 2022 and 2023 batter stats
df_batting_HitByPitch = [df_batting_2023['Times_Hit_By_Pitch'].sum(), df_batting_2022['Times_Hit_By_Pitch'].sum()]

# Creates a list of years to be used in the creation of certain graphs below
years = ['2022', '2023']
```

```python
figure, axs = plt.subplots(1, 4, figsize=(15, 5)) # Defines the figure and subplots in the figure. 1 row and 4 columns for the figure

# Creates the first subplot graph that shows the average age of players. Bar graph here
axs[0].bar(years, df_batting_Age, color = 'purple') 
axs[0].set_title('Average Age of Players')

# Creates the second subplot graph that shows the total number of hits for each season. Bar graph here
axs[1].bar(years, df_batting_hits, color = 'red')
axs[1].set_title('Total Number of Hits')

# Creates the third subplot graph that shows the total number of home runs for each season. Bar graph here
axs[2].bar(years, df_batting_HomeRuns, color = 'green')
axs[2].set_title('Total Number of Home Runs')

# Creates the second subplot graph that shows the total number of times hit by pitch for each season. Bar graph here
axs[3].bar(years, df_batting_HitByPitch, color = 'yellow')
axs[3].set_title('Total Times Hit by Pitch')
```

```python
fig, ax2 = plt.subplots() # Creates a figure and a subplot for the following scatterplot


# Creates the scatterplot of from the 2023 batting dataframe. Shows data for the number of at_bats and hits occurred for the first 5 rows
ax2.scatter(df_batting_2023['At_Bats'].head(), df_batting_2023['Hits'].head(), marker='o', color = 'blue')

# Sets the title of the graph
ax2.set_title('At Bats vs Hits for First Five Batters')

# Sets the grid to be visible in the graph
ax2.grid(True)

# Sets the xlabel and ylabels
ax2.set_xlabel('At Bats')
ax2.set_ylabel('Hits')
```
