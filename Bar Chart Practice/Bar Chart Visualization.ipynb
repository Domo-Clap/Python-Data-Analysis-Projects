```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
```

```python
df_used_cars = pd.read_csv("bar_chart_data.csv")
sea.set()
```

```python
df_used_cars
```

```python
plt.figure(figsize = (9, 6))

plt.bar(x = df_used_cars["Brand"], height = df_used_cars["Cars Listings"], color = "violet")
plt.xticks(rotation = 45)
#plt.title("Car Listings by Brand", fontsize = 16, fontweight = "bold")
plt.ylabel("Number of Listings", fontsize=13)
plt.yticks(fontsize=12)

plt.show()
```
