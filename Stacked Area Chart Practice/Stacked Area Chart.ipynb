```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
```

```python
df_fuel_engine_types = pd.read_csv("stacked_area_chart_data.csv")
df_fuel_engine_types
```

```python
colors = ["#011638", "#7e2987", "#ef2026"]
labels = ["Diesel", "Petrol", "Gas"]

plt.figure(figsize = (12, 6))

plt.stackplot(df_fuel_engine_types["Year"], df_fuel_engine_types["Diesel"], df_fuel_engine_types["Petrol"], df_fuel_engine_types["Gas"], colors = colors)

plt.xticks(df_fuel_engine_types["Year"], rotation = 45)
plt.legend(labels = labels, loc = "upper left")
plt.ylabel("Number of Cars")
plt.title("Popularity of Engine Fuel Types", fontweight = "bold")
sea.despine()
plt.show()
```
