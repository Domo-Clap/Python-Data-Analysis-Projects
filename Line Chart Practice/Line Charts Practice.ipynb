```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
seaborn.set()
```

```python
df_stock_market = pd.read_csv("line_chart_data.csv")
df_stock_market
```

```python
df_stock_market["new_date"] = pd.to_datetime(df_stock_market["Date"])
```

```python
df_stock_market
```

```python
labels = ["S&P 500", "FTSE 100"]

plt.figure(figsize = (20, 8))

plt.plot(df_stock_market["new_date"], df_stock_market["GSPC500"])
plt.plot(df_stock_market["new_date"], df_stock_market["FTSE100"])
plt.title("S&P vs. FTSE Returns", fontweight = "bold", fontsize = 16)

plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = "large")

plt.show()
```

```python
df_stock_market_2008 = df_stock_market[(df_stock_market.new_date >= '2008-07-01') & (df_stock_market.new_date <= '2008-12-31')]
df_stock_market_2008
```

```python
labels = ["S&P 500", "FTSE 100"]

plt.figure(figsize = (18, 7))

plt.plot(df_stock_market_2008["new_date"], df_stock_market_2008["GSPC500"])
plt.plot(df_stock_market_2008["new_date"], df_stock_market_2008["FTSE100"])
plt.title("S&P vs. FTSE Returns", fontweight = "bold", fontsize = 16)

plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = "large")

plt.grid(True)
plt.show()
```
