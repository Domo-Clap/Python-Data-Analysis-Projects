```python
sea.set_style(style='dark') # Sets the style for the chart to dark mode
sea.countplot(x='City', data=df).set_title('City Count') # Creates a histogram like graph showing the count of cities in the data frame
```

```python
sea.countplot(x='Gender', data=df).set_title('Genders') # Creates a histogram like graph showing the count of genders in the data frame
```

```python
sea.displot(x= 'Rating' ,data=df) # Creates a distribution chart showing the amount of each rating within the dataframe. Default histogram mode here
```

```python
sea.displot(x= 'Rating' ,data=df, kind='ecdf') # Shows the above distribution chart in a different form. Proportion distribution functions used for this one. Not using the base histogram mode
```

```python
sea.displot(x='cogs', data=df, kde = True) # Shows the above distribution chart in a different form. Default Histogram mode here but with a KDE being used.
```

```python
sea.scatterplot(data = df, x='Unit price', y='Rating') # Scatterplot used to show the comparison between the costs of certain products and their reviews
```

```python
sea.barplot(data=df, x = 'Branch', y='gross income') # Creates a barplot showing how much each branch made in terms of gross income
```

```python
sea.barplot(data=df, x = 'Product line', y='Unit price') # Creates a barplot showing the different product line types and the highest unit price for each
```
